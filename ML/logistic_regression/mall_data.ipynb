```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
#We will take Age and Estimated salary as features and Purchased as the label

X = dataset.iloc[: , [2,3]]
y = dataset.iloc[: , 4]
```

```python
print(X.shape)
print(y.shape)
```

```python
#Now we will divide the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.25, random_state=0)
```

```python
#Since we have features such as age and salary which could differ too much, therefore we will scale it in a uniform range
# 0-1
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
X_test
```

```python
X_train
```

```python
#Creating the model and fitting it to our traning set
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
```

```python
#predicting the results
y_pred = clf.predict(X_test)
```

```python
y_pred.shape
```

```python
#Evaluating the performance of the model using confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
